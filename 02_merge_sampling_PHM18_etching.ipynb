```python
# 라이브러리 불러오기

import pandas as pd
import os
import numpy as np
import glob
```

```python
dir_train = '/content/drive/MyDrive/phm18/train'
dir_ttf = '/content/drive/MyDrive/phm18/train_ttf'

files_train = sorted( glob.glob(os.path.join(dir_train, '*csv')) ) # '*csv'를 통해 확장자가 .csv인 파일만 골라내기
files_ttf = sorted( glob.glob(os.path.join(dir_ttf, '*csv')) )

print(f"train 파일 개수: {len(files_train)}")
print(files_train)
print()
print(f"train_ttf 파일 개수: {len(files_ttf)}")
print(files_ttf)
```

```python
# 컬럼 정의하기
sensor_cols = [
    'IONGAUGEPRESSURE', 'ETCHSUPPRESSORCURRENT', 'FLOWCOOLPRESSURE',
    'ACTUALROTATIONANGLE', 'ACTUALSTEPDURATION'
]

control_cols = [
    'ETCHBEAMVOLTAGE', 'ETCHBEAMCURRENT', 'ETCHSUPPRESSORVOLTAGE',
    'FLOWCOOLFLOWRATE', 'ETCHGASCHANNEL1READBACK', 'ETCHPBNGASREADBACK',
    'FIXTURETILTANGLE', 'ROTATIONSPEED', 'FIXTURESHUTTERPOSITION',
    'ETCHSOURCEUSAGE', 'ETCHAUXSOURCETIMER', 'ETCHAUX2SOURCETIMER'
]

ttf_cols = [
    'TTF_pressure_drop', 'TTF_pressure_high', 'TTF_leak'
]
```

```python
# 데이터 확인

df_01 = pd.read_csv(files_train[0])
print(df_01.shape)
print(df_01.columns.tolist())
```

```python
dfs = [] # 총 20개의 데이터를 담을 바구니 만들기

for file_train, file_ttf in zip(files_train, files_ttf):
  df_train = pd.read_csv(file_train)
  df_ttf = pd.read_csv(file_ttf)
  df_merged = df_train.merge(df_ttf, on='time')
  df_sampled = df_merged.sample(frac=0.01, random_state=42) # 합친 데이터에서 1%만 랜덤 추출, random_state (중요)
  dfs.append(df_sampled) # sampled 결과를 바구니에 넣기
  print(f"{file_train} 처리 완료: {df_sampled.shape}")

df_all_1pct = pd.concat(dfs, axis=0, ignore_index=True) # ignore_index=True는 행 번호를 0부터 새로 매겨줌
print(df_all_1pct.shape)
```

```python
# 매번 샘플링하기 번거로우므로, 샘플링 된 결과를 저장해두자

df_all_1pct.to_csv('/content/drive/MyDrive/phm18/df_all_sampled_1pct.csv', index=False)
# index=False는 나중에 데이터를 불러올 때 불필요한 컬럼이 생기지 않게 해줌
print("저장 완료!", df_all_1pct.shape)
```
